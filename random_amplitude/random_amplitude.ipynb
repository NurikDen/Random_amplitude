```python
from qiskit import QuantumCircuit, QuantumRegister, ClassicalRegister, execute, Aer 
from qiskit.quantum_info import random_statevector 
from qiskit.visualization import plot_histogram 
import numpy as np 
state = random_statevector(4).data
qr = QuantumRegister(2)
cr = ClassicalRegister(2) 
circuit = QuantumCircuit(qr, cr) 
circuit.initialize(state, [0, 1])
circuit.measure(qr, cr)
simulator = Aer.get_backend('qasm_simulator') 
job = execute(circuit, simulator, shots=100) 
result = job.result() 
counts = result.get_counts()
max_perc = max(counts, key=counts.get)
value = int(max_perc, 2)
print("Состояние:", max_perc)
print("Эквивалент в десятичной системе:", value)
plot_histogram(counts)
#Перед запуском нужно установить библиотеку qiskit и numpy (желательно запускать в jupyter notebook)
```
